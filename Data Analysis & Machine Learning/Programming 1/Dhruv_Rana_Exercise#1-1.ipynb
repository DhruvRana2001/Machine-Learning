### Problem 1.3

```python
fetal_dirty_df['baseline value'] = fetal_dirty_df['baseline value'].where(fetal_dirty_df['baseline value']>0,np.NaN)
fetal_dirty_df['percentage_of_time_with_abnormal_long_term_variability'] = fetal_dirty_df['percentage_of_time_with_abnormal_long_term_variability'].where(fetal_dirty_df['percentage_of_time_with_abnormal_long_term_variability']>=0,np.NaN)
fetal_dirty_df['percentage_of_time_with_abnormal_long_term_variability'] = fetal_dirty_df['percentage_of_time_with_abnormal_long_term_variability'].where(fetal_dirty_df['percentage_of_time_with_abnormal_long_term_variability']<=100,np.NaN)
fetal_dirty_df['histogram_width'] = fetal_dirty_df['histogram_width'].where(fetal_dirty_df['histogram_width']>0,np.NaN)

fetal_dirty_df.loc[91:100]
```

_The baseline value is the heart rate of the fetus and cannot be negative thus all negative heart rates and 0 heart rates were turned to NaN. Also, the percentage cannot be lower than 0 and higher than 100 and must be converted to NaN. At last, the width of the histogram cannot be 0._

### Problem 1.4

```python
# Mean Normalization
for col in fetal_dirty_df.columns[:-2] :
    fetal_dirty_df[col] = (fetal_dirty_df[col] - fetal_dirty_df[col].mean())  / fetal_dirty_df[col].std()

fetal_dirty_df.head(20)
# even tough mean is not exacctly zero the number are so small that they can be estimated to be zero
```

### Problem 1.5

```python
fetal_dirty_df.fillna(fetal_dirty_df.median(), inplace = True)
```

### Problem 1.6

```python
#pd.concat([fetal_dirty_df,pd.get_dummies(fetal_dirty_df.fetal_health,prefix = "fetal_health")],axis=1).drop(['fetal_health'],axis=1)

# 1 -  Normal  -> 100
# 2 - Suspect  -> 010
# 3 - Pathological -> 001

fetal_h = []

for i in range(len(fetal_dirty_df)):
    if fetal_dirty_df.loc[i,'fetal_health'] == 1 :
        fetal_h.append(np.array([1,0,0]))
    elif fetal_dirty_df.loc[i,'fetal_health'] == 2 :
        fetal_h.append(np.array([0,1,0]))
    elif fetal_dirty_df.loc[i,'fetal_health'] == 3 :
        fetal_h.append(np.array([0,0,1]))
    
fetal_dirty_df.drop(['fetal_health'],axis =1)
fetal_dirty_df['fetal_health'] = fetal_h

fetal_dirty_df.to_csv('fetal_health_dataset_processed.csv', index=False)
```

## Problem 2.2

```python
heart_df['age'] = heart_df['age'].where(heart_df['age'] >= 0, np.NaN)
heart_df['ejection_fraction'] = heart_df['ejection_fraction'].where(heart_df['ejection_fraction'] >= 0, np.NaN)
heart_df['ejection_fraction'] = heart_df['ejection_fraction'].where(heart_df['ejection_fraction'] <= 100, np.NaN)
```

_We checked logical consitency in age and ejection fraction. Age cannot be a neagtive number, and ejection fraction is a percentage and cannot be grater than 100 or smaller than 0._

## Problem 2.3

```python
for col in heart_df.columns[:] :
    if not (col.__eq__("sex") | col.__eq__("smoking") | col.__eq__("anaemia") | col.__eq__("diabetes") | col.__eq__("high_blood_pressure") | col.__eq__("DEATH_EVENT")):
        heart_df[col] = (heart_df[col] - heart_df[col].mean())  / heart_df[col].std()

heart_df.head(20)
```

## Problem 2.4

```python
# Deceased  patients -> 1 -> 10 
# Remaining patients -> 0 -> 01
heart_df1 = heart_df.copy(deep=False)
death = []

for i in range(len(heart_df1)):
    if heart_df1.loc[i,'DEATH_EVENT'] == 1 :
        death.append(np.array([1,0]))
    elif heart_df1.loc[i,'DEATH_EVENT'] == 0 :
        death.append(np.array([0,1]))
    
heart_df1.drop(['DEATH_EVENT'],axis =1)
heart_df1['DEATH_EVENT'] = death

heart_df1.to_csv(' heart_failure_dataset_processed.csv', index=False)
```
